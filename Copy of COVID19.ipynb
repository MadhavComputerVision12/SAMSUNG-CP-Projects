```python
path='/content/drive/MyDrive/nation_level_daily.csv'
data=pd.read_csv(path)
```

```python
print(data.head(10))
```

```python
sns.lineplot(data['Daily Confirmed'])
```

```python
n=25
day_average=data['Daily Confirmed']
moving_average=[0]*len(day_average)
for i in range(len(day_average)):
  if(i==0):
    moving_average[i]=day_average[i]
  elif(i<n and i!=0):
    moving_average[i]=(sum(day_average[0:i])/i)
  else:
    moving_average[i]=sum(day_average[i-n:i])/n
print(moving_average)
```

```python
plt.plot(data['Daily Confirmed'],color='black')
plt.plot(moving_average,color='red')
plt.show()
#25 day moving average
```

```python
#seeing the graph of deceased / total new cases ratio
ratio=data['Total Deceased']/data['Total Confirmed']
print(ratio)
```
