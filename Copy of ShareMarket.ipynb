```python
path = "/content/drive/MyDrive/SKX.csv"
data = pd.read_csv(path)
```

```python
plt.plot(data["High"])
```

```python
#considering the average on each day is average of high and low
day_average=list((data["High"]+data["Low"]+data["Open"]+data["Close"])/4.0)
```

```python
print(day_average)
```

```python
#taking the user input
n=int(input("Enter the n to plot n-day moving average : "))
```

```python
moving_average=[0]*len(day_average)
for i in range(len(day_average)):
  if(i==0):
    moving_average[i]=day_average[i]
  elif(i<n and i!=0):
    moving_average[i]=(sum(day_average[0:i])/i)
  else:
    moving_average[i]=sum(day_average[i-n:i])/n
print(moving_average)
```

```python
plt.plot(data["High"])
plt.plot(moving_average)
```

```python
#calculate the error between two
error=[]
for i in range(len(moving_average)):
  error.append(abs(((day_average[i]-moving_average[i])/day_average[i])*100))
print(error)
plt.scatter(data['Volume'],y=error)
```

```python
def fibonacci_levels(i,arr):
  x=max(arr[0:i+1])
  l=[]
  l.append(0.236*x)
  l.append(0.382*x)
  l.append(0.618*x)
  return l
```

```python
#visulisation of fibonacci supports of moving averages
l=[]
final_1=[]
final_2=[]
final_3=[]
for i in range(len(day_average)):
  l.clear()
  l=fibonacci_levels(i,data['High'])
  final_1.append(l[0])
  final_2.append(l[1])
  final_3.append(l[2])
plt.plot(data['High'],color='black')
plt.plot(final_1,color='red')
plt.plot(final_2,color='blue')
plt.plot(final_3,color='green')
plt.show()
```

EMA=Price(t)×k+EMA(y)×(1−k)
where:
t=today
y=yesterday
N=number of days in EMA
k=2÷(N+1)
​


```python
#exponential moving average
#here price(t) is replaced by average price of day
n=int(input("Enter the value for number of days : "))
```

```python
day_new=day_average[n:]
ema=[0]*(len(day_average)-n)
for i in range(0,len(day_new)):
  k=2/(n+1)
  ema[i]=day_new[i]*k+ema[i-1]*(1-k)
```

```python
plt.plot(ema,color='red')
plt.plot(day_new,color='black')
plt.show()
```
